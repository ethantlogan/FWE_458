```python
#This code allows access to my google drive and makes it so pandas can read
#the file with the data and import it into a dataframe in google colab
from google.colab import drive
drive.mount('/content/drive')

filedir = '/content/drive/MyDrive/FWE458/'
filename = filedir + "MaunaLoaCO2_monthly_sitename.csv"

import pandas as pd
DF = pd.read_csv(filename)
DF
```

```python
#This adds two filters by defining the date frame to include only values from
#the "Year" column that are greater than 1958 or less than 2022
#Then it creates a new data frame using the filters created from the original
#data frame
filter1 = DF["Year"] > 1958
filter2 = DF["Year"] < 2022
DF_new = DF[filter1 & filter2]
DF_new
```

```python
#This is another way to create the filters by just defining the new data frame
#as the original date frame without the years we want to exclude
DF_new = DF[(DF["Year"] > 1958) & (DF["Year"] < 2022)]
DF_new
```

```python
#This is creating more filters to get rid of all negative numbers in the two
#columns we want to filter. It makes them include both 0 and positive numbers
#Then it creates a new data frame using the two new filters
filter3 = DF_new["numofdays"] >= 0
filter4 = DF_new["stdofdays"] >= 0
DF_new2 = DF_new[filter3 & filter4]
DF_new2
```

```python
#This defines a new list as DF_new3 as the mean of the monthly CO2 concentrations
#The new list is created by first grouping the data from the filtered data into
#year categories, then applying a "mean" function to the Monthly_average
#column and finding the mean of those values in each grouped category.
import pandas as pd
import numpy as np
DF_new3 = DF_new2.groupby("Year")["Monthly_average"].mean()
DF_new3
```

```python
#This creates a new column for the annual average CO2 concentrations and makes
#it so the new column is filled in with the grouped data from the mean
#calculations done before. The transform function allows the data to be put in
#all the rows, and assigns the value in those rows based off the year category
#it's in. It calculates the mean for each group of "Year" and based off of the
#numbers from Monthly_average and assigns that new mean to every member in the
#group.
DF_new2["annual_average"] = DF_new2.groupby("Year")["Monthly_average"].transform("mean")
DF_new2
```

```python
#This does the exact same thing as above but with new columns and numbers
DF_new2["deseasonalized_annual_average"] = DF_new2.groupby("Year")["deseasonalized"].transform("mean")
DF_new2
```

```python
#This again does the same as above but with different columns, numbers, and
#functions. So this uses the standard deviation (std) function instead of the
#mean function to calculate the new numbers to be assigned.
DF_new2["std_of_months"] = DF_new2.groupby("Year")["Monthly_average"].transform("std")
DF_new2
```

```python
#This created a function that evaluates the numbers from the Monthly_average
#column and then sorted those numbers into 3 groups based on their value.
#Then I created a new column in the data frame, and added the new assignment
#based on the CO2 concentration number using the transform command which
#followed my function that I created to assign the "Low"/"Moderate"/"High"
#to each value. The transfrom made sure each value in the index got an
#assignment and a value in the new column created. This code also ensured that
#the numbers used for the assignments came from the Montly_average column.
def cat(Monthly_average):
    if Monthly_average < 350:
      return("Low")
    elif Monthly_average <= 400:
      return("Moderate")
    else:
      return("High")

DF_new2["CO2_category"] = DF_new2["Monthly_average"].transform(cat)
DF_new2
```
